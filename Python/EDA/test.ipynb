```python
# Step 1: Business Understanding and Step 2: Domain Knowledge
# These steps involve understanding the business context and the domain. Document this separately in markdown cells or notes.

# Step 3: Data Cleaning

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('test.csv', delimiter=';')  # Updated to use the correct delimiter based on your data

# Check for missing values
print("Missing values in each column:\n", df.isnull().sum())

# Display basic information about the dataset (data types, non-null counts, etc.)
print("\nBasic info of the dataset:\n")
df.info()

# Display unique values in categorical columns to check for consistency
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    print(f'\nUnique values in {col}:\n{df[col].unique()}')

# Check for basic statistics of numeric columns
print("\nSummary statistics of numeric columns:\n")
print(df.describe())

# Optional: Check for outliers in numerical data (visual inspection can also be done via box plots)
sns.boxplot(data=df.select_dtypes(include=['int64', 'float64']))
plt.title('Boxplot of Numeric Features')
plt.show()

# Step 4: Data Analysis

# Distribution of the target variable (y)
plt.figure(figsize=(6, 4))
sns.countplot(x='y', data=df, palette='Set2')
plt.title('Distribution of Target Variable (Subscribed to Term Deposit)')
plt.xlabel('Subscribed to Term Deposit (y)')
plt.ylabel('Count')
plt.show()

# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], kde=True, bins=20, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Box plot for balance distribution by subscription status
plt.figure(figsize=(8, 6))
sns.boxplot(x='y', y='balance', data=df, palette='Set3')
plt.title('Balance Distribution by Subscription Status')
plt.xlabel('Subscribed to Term Deposit')
plt.ylabel('Balance')
plt.show()

# Count of subscription by job
plt.figure(figsize=(10, 6))
sns.countplot(y='job', hue='y', data=df, palette='Set1')
plt.title('Term Deposit Subscription by Job Type')
plt.xlabel('Count')
plt.ylabel('Job Type')
plt.show()

# Relationship between age and balance
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='balance', hue='y', data=df, palette='coolwarm', alpha=0.6)
plt.title('Age vs Balance (Colored by Subscription Status)')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.show()

# Step 4: Updated Correlation Matrix for Numerical Features

# Filter the dataframe to include only numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Compute the correlation matrix
plt.figure(figsize=(10, 8))
corr_matrix = numeric_df.corr()  # Only numeric columns are passed
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# Step 5: Optional Visualization - Pair Plot for numerical features
sns.pairplot(df, hue='y', diag_kind='kde', markers=["o", "s"], palette="Set2")
plt.show()

```
